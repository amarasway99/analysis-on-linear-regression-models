```python
import pandas as pd
import numpy as np

# Selecting rows 2 to 41 for X and y
X_subset = df.iloc[0:40, :-1].values  # Select all columns except the last one (which is 'y')
y_subset = df.iloc[0:40, -1].values  # Select the 'y' column

"""
# Verifying the output
print("X_subset:", X_subset[:5])  # Print first 5 rows of X_subset to verify
print("y_subset:", y_subset[:5])  # Print first 5 rows of y_subset to verify
print("X_subset:", X_subset[-5:])  # Print first 5 rows of X_subset to verify
print("y_subset:", y_subset[-5:])  # Print first 5 rows of y_subset to verify
"""

# Compute X'X
X_transpose = np.transpose(X_subset)
XtX = np.dot(X_transpose, X_subset)

# Compute (X'X)^-1
XtX_inv = np.linalg.inv(XtX)

# Compute X'y
Xty = np.dot(X_transpose, y_subset)

# Perform the final matrix computation ((X'X)^-1 * X'y)
result = np.dot(XtX_inv, Xty)

# Display the result
result
```

```python
# Selecting rows 2 to 41 for X and y
X_subset = df.iloc[0:40, :-1].values  # Select all columns except the last one (which is 'y')
y_subset = df.iloc[0:40, -1].values  # Select the 'y' column

# Select calculated w values from Question #1
w = result

# Generate predicted y values from w
y_predicted_train = np.dot(X_subset, w)

# Computer SSR for training data
ssr_train = np.sum((y_subset - y_pred_train) ** 2)

# Computer MSR for training data
msr_train = ssr_train / len(y_subset)

# Compute RMSE for training data
rmse_train = np.sqrt(msr_train)

print(f"""
Training Errors:
----------------
SSR  : {ssr_train:.4f}
MSR  : {msr_train:.4f}
RMSE : {rmse_train:.4f}
""")
```

```python
# Selecting rows 43 to 51 for test data
X_test = df.iloc[41:50, :-1].values  # Select all columns except the last one from index 41 to 50
y_test = df.iloc[41:50, -1].values  # Select 'y' column from index 41 to 50

"""
# Verifying the output
print("X_test:", X_test[:5])  # Print first 5 rows of X_subset to verify
print("y_test:", y_test[:5])  # Print first 5 rows of y_subset to verify
print("X_test:", X_test[-5:])  # Print first 5 rows of X_subset to verify
print("y_test:", y_test[-5:])  # Print first 5 rows of y_subset to verify
"""

# Generate predicted y values for test data
y_pred_test = np.dot(X_test, w)

# Compute SSR for test data
ssr_test = np.sum((y_test - y_pred_test) ** 2)

# Compute MSR for test data
msr_test = ssr_test / len(y_test)

# Compute RMSE for test data
rmse_test = np.sqrt(msr_test)

# Print the Test Errors
print(f"""
Test Errors:
----------------
SSR  : {ssr_test:.4f}
MSR  : {msr_test:.4f}
RMSE : {rmse_test:.4f}
""")
```

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split

# Load the data
file_path = "GroupExercise-2.xlsx"
sheet_name = "Data"
df = pd.read_excel(file_path, sheet_name = sheet_name)

#Split the data Rows 2 to 11 for training, 12 to 51 for testing
train = df.iloc[0:10] 
test = df.iloc[10:51]
# Verifying the output
print("train:", train[:])
print("test:", test[:])  


# Features (X) and target (y) 
X_train = train[['X0', 'X1', 'X2', 'X3', 'X4', 'X5']]
y_train = train['y']

X_test = test[['X0','X1' ,'X2', 'X3', 'X4', 'X5']]
y_test = test['y']

# Train the Ridge Regression model with lambda = 0.17
lambda_value = 0.17
model = Ridge(alpha=lambda_value)
model.fit(X_train, y_train)

# Make predictions
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)

# Evaluate the model
mse_train = mean_squared_error(y_train, y_train_pred)
mse_test = mean_squared_error(y_test, y_test_pred)
# Calculate R-squared (R²) for training and testing sets
r2_train = r2_score(y_train, y_train_pred)
r2_test = r2_score(y_test, y_test_pred)

# Return the results: MSE on train, MSE on test, R² on train, and R² on test
print("MSE on train: \n", mse_train)
print("MSE on test: \n", mse_test)
print("R² on train: \n", r2_train)
print("R² on test: \n", r2_test)
```
